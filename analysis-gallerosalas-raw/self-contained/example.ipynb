```python
import numpy as np
import matplotlib.pyplot as plt
from brainplot_selfcontained import load_matrix, plot_area_values, plot_area_clusters
```

```python
allenMap, allenIndices = load_matrix('')
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(8, 4), tight_layout=True)

# Figure 1: We will plot random real numbers from 0 to 1 for each region
values1 = np.random.uniform(0, 1, 27)   # There are 27 regions
plot_area_values(fig, ax[0], allenMap, allenIndices, values1, vmin=0, vmax=1, haveColorBar=True)

# Figure 2: We will have data from 3 different classes (I just put some random areas here)
values2 = {
    'class1' : [0, 1, 2, 3, 4, 5, 6, 7],
    'class2' : [10, 11, 12, 15, 16, 20],
    'class3' : [13, 22, 23]
}
plot_area_clusters(fig, ax[1], allenMap, allenIndices, values2, haveLegend=True, haveColorBar=True)

plt.show()
```
